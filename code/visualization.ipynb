```python
from utils import create_gdf_from_names
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap


gdf = create_gdf_from_names(["magdeburg", "berlin"]) 

fig = plt.figure(figsize = (12,12))
m = Basemap()
m.drawcoastlines(linewidth=1.0, linestyle='solid', color='black')
m.drawcountries(linewidth=1.0, linestyle='solid', color='k')
m.fillcontinents()

lons = gdf.lon.tolist()
lats = gdf.lat.tolist()


x, y = m(lons, lats)

m.scatter(x, y)

plt.title("Color filled continents", fontsize=20)
plt.show()



print(gdf.lon.tolist())

           
```
