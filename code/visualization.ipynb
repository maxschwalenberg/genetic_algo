```python
from utils import create_gdf_from_names, get_basemap_params_from_gdf
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap


gdf = create_gdf_from_names(["magdeburg", "berlin", "rostock", "kopenhagen", "madrid"]) 
lower_left, upper_right, mid = get_basemap_params_from_gdf(gdf, 2)


fig = plt.figure(figsize = (12,12))
m = Basemap(llcrnrlon=lower_left[0], llcrnrlat=lower_left[1], urcrnrlon=upper_right[0], urcrnrlat=upper_right[1])
m.drawcoastlines(linewidth=1.0, linestyle='solid', color='black')
m.drawcountries(linewidth=1.0, linestyle='solid', color='k')
m.fillcontinents()

lons = gdf.lon.tolist()
lats = gdf.lat.tolist()


x, y = m(lons, lats)

m.scatter(x, y)

plt.title("Color filled continents", fontsize=20)
plt.show()



print(gdf.lon.tolist())

           
```
